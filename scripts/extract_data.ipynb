```python
df = pd.read_excel("../data/exp_data/PDBbind_complete_general_set_exp.xlsx",header=1)

print(df.keys())

df_ec = df[['PDB code','EC number']]

df_ec

ec_1 = []
ec_2 = []
ec_3 = []
ec_4 = []


not_unique = []
for n, j in enumerate(df_ec["EC number"]):
    b = j.split(";")
    if len(b)>1:
        not_unique.append(n)
        
df_ec.drop(not_unique,axis="index",inplace=True)
df_ec.reset_index(inplace=True,drop=True)
print(len(not_unique))

for i in df_ec["EC number"]:
    a = i.split(".")
    ec_1.append(a[2])
    try:
        ec_2.append(a[3])
    except:
        ec_2.append("-")
    try:
        ec_3.append(a[4])
    except:
        ec_3.append("-")
    try:
        ec_4.append(a[5])
    except:
        ec_4.append("-")
        
print(len(ec_1),len(ec_2),len(ec_3),len(ec_4))
df_ec["Class"] = ec_1
df_ec["Subclass"] = ec_2
df_ec["Subsubclass"] = ec_3


remove = []
for n, m in enumerate(df_ec["Class"]):
    try:
        tmp = int(m)
        if tmp > 10:
            remove.append(n)
    except:
        remove.append(n)
        
df_ec.drop(remove,axis="index",inplace=True)
df_ec.reset_index(inplace=True,drop=True)


remove = []
for n, m in enumerate(df_ec["Subclass"]):
    if m == "-":
        remove.append(n)
        
df_ec.drop(remove,axis="index",inplace=True)
df_ec.reset_index(inplace=True,drop=True)
df_ec.to_csv("../data/exp_data/EC_numbers/ec_numbers_all.csv")
```

```python
unique_classes = []
for i in df_ec["Class"]:
    if int(i) in unique_classes:
        continue
    else:
        unique_classes.append(int(i))
unique_classes.sort()
print("Classes",unique_classes)


for c in unique_classes:
    df_tmp = df_ec.loc[df_ec["Class"] == str(c)]
    df_tmp.to_csv(f"../data/exp_data/EC_numbers/{c}_x_x_x.csv")
    print(f"Dataframe for class {c} has shape:",df_tmp.shape)
```

```python
descriptors = ["X-GRADE","GRADE"]
Classes = ["1","2","3","4","5","6","7"]

for des in descriptors:
    test_grail = f"../data/Descriptors/PDBbind_general_set_{des}.csv"
    val_grail = f"../data/Descriptors/PDBbind_core_set_{des}.csv"
    test_exp = "../data/exp_data/PDBbind_general_set_all.csv"
    val_exp = "../data/exp_data/PDBbind_core_set_all.csv"

    df_test_grail = pd.read_csv(test_grail)
    df_val_grail = pd.read_csv(val_grail)
    df_test_exp = pd.read_csv(test_exp)
    df_val_exp = pd.read_csv(val_exp)
    for c in Classes:
        df_class = pd.read_csv(f"../data/exp_data/EC_numbers/{c}_x_x_x.csv",index_col=0)
        PDBs = df_class["PDB code"]
        df_test_grail_class = df_test_grail[df_test_grail["PDB code"].isin(PDBs)]
        df_val_grail_class = df_val_grail[df_val_grail["PDB code"].isin(PDBs)]
        df_test_exp_class = df_test_exp[df_test_exp["PDB code"].isin(PDBs)]
        df_val_exp_class = df_val_exp[df_val_exp["PDB code"].isin(PDBs)]
        df_test_grail_class.to_csv(f"../data/Descriptors/EC_numbers/PDBbind_general_set_{des}_class{c}.csv",index=False)
        df_val_grail_class.to_csv(f"../data/Descriptors/EC_numbers/PDBbind_core_set_{des}_class{c}.csv",index=False)
        df_test_exp_class.to_csv(f"../data/exp_data/EC_numbers/general_set_class{c}.csv",index=False)
        df_val_exp_class.to_csv(f"../data/exp_data/EC_numbers/core_set_class{c}.csv",index=False)
        print(f"Class {c} done")
```
